# Cleaning the data <br>   
1-Classification of types of columns<br>
2-Handling empty and duplicate values<br>

 

```python
df.info()
```

```python
df['session_date'] = pd.to_datetime(df['session_date'])
df['signup_date'] = pd.to_datetime(df['signup_date'])
df['last_login_date'] = pd.to_datetime(df['last_login_date'])

```

```python
df.sample(5)
```

**Edit login date**

```python

invalid_rows = df[df['session_date'] < df['signup_date']]
p = (len(invalid_rows) / len(df)) * 100

print(f"Number of incorrect rows : {len(invalid_rows)}")
print(f" percentage : {p}%")
```

```python
df['signup_date'] = df[['signup_date', 'session_date']].min(axis=1)
df['signup_date'] = df.groupby('user_id')['signup_date'].transform('min')

invalid_rows = len(df[df['session_date'] < df['signup_date']])
f"Number of incorrect rows : {invalid_rows}"
```

# Feature Extraction

```python
df.sample(5)
```

```python
df['Age Group'] = pd.cut(df['age'], 
                        bins=[0, 23, 45, 100], 
                        labels=['Young', 'Mid', 'Adult'])

df['sub_duration_days'] = (df['last_login_date'] - df['signup_date']).dt.days
df['goal_reached_numeric'] = df['completed_goal_checkin'].map({'Yes': 1, 'No': 0})

```

# Analysis


```python
unique_users_count = df['user_id'].nunique()
total_sessions = len(df)
avg_sessions_per_user = (total_sessions / unique_users_count)

print(f"Total Unique Users: {unique_users_count}")
print(f"Total Sessions in Dataset: {total_sessions}")
print(f"Average Sessions per User: {int(avg_sessions_per_user)}")
```

## 1 - Percentage of men and women<br>
## 2 - Percentage of each age group

```python
gender_counts = df.groupby('user_id')['gender'].first().value_counts()
gender_counts
```

```python
plt.figure(figsize=(8, 5))
colors = ['#A8DADC', '#E76F51'] 
plt.pie(gender_counts
        ,labels=['Male','Female'],autopct='%1.1f%%',shadow=True,
        startangle=140,pctdistance=0.85, 
        explode=(0.05, 0) ,colors=colors)
plt.title('Gender Distribution', fontsize=15)
plt.axis('equal')
plt.show()
```

```python
age_counts = df.groupby('user_id')["Age Group"].first().value_counts()
age_counts
```

```python
plt.figure(figsize=(8, 5))
colors = ['#CAF0F8', '#90E0EF', '#00B4D8']
plt.pie(age_counts
        ,labels=age_counts.index,autopct='%1.1f%%',shadow=True,
        startangle=140,pctdistance=0.85, 
        colors=colors)
plt.title('Age Group Distribution', fontsize=15)
plt.axis('equal')
plt.show()
```

## Number of active and inactive subscribers and its number in each age group

```python
dropoff_counts = df.groupby('user_id')['dropoff_flag'].first().value_counts()
dropoff_counts
```

```python
plt.figure(figsize=(8, 5))
colors = ['#CAF0F8', '#00B4D8'] 
plt.pie(dropoff_counts
        ,labels=['Active','NOT Active'],autopct='%1.1f%%',shadow=True,
        startangle=140,pctdistance=0.85, 
        explode=(0.05, 0) ,colors=colors)
plt.title('Dropoff flag Distribution', fontsize=15)
plt.axis('equal')
plt.show()
```

```python
not_active_df = df.groupby('user_id').first()
not_active_df = not_active_df[not_active_df['dropoff_flag'] == 'Yes']
dropoff_by_age = not_active_df['Age Group'].value_counts().reindex(['Young', 'Mid', 'Adult'])

dropoff_by_age.plot(kind='bar', color="#457B9D", edgecolor='black')

plt.title('Distribution of NOT Active Users (Drop-off) by Age Group', fontsize=15)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Number of Users', fontsize=12)
plt.xticks(rotation=0)
plt.grid(linestyle='--', alpha=0.5)

plt.show()
```

```python
train_types = df['workout_module_used'].unique()
print("Training Types Available:")
print(train_types)
```

```python
workout_counts = df['workout_module_used'].value_counts()

workout_counts.plot(kind="bar" , color="#457B9D")
plt.title('Distribution of Workout Modules')
plt.xlabel('Workout Type')
plt.ylabel('Number of Sessions')
plt.xticks(rotation=45)
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
workout_gender_counts = pd.crosstab(df['workout_module_used'], df['gender'])

workout_gender_counts.plot(kind='bar', figsize=(8, 6), color=["#457B9D", "#46494A"]) 

plt.title('Workout Preferences by Gender', fontsize=15)
plt.xlabel('Workout Module', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Gender') 
plt.grid(axis='y', linestyle='--', alpha=0.6) 

plt.tight_layout()
plt.show()
```

```python
age_workout_calories = df.pivot_table(index='workout_module_used', 
                                    columns='Age Group', 
                                    values='calories_burned', 
                                    aggfunc='mean')
age_workout_calories.plot(kind='bar', figsize=(10, 7), color=["#B5CBCC", '#457B9D', "#0C2A3C"])
plt.title('Average Calories Burned by Workout Type and Age Group', fontsize=15)
plt.xlabel('Workout Module', fontsize=12)
plt.ylabel('Average Calories Burned', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Age Group', labels=['Young (0-23)', 'Mid (24-45)', 'Adult (46+)'])
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

**1 - The relationship between each training system and the user's objective**<br>
**2 - The goal of the training for both men and women**<br>
**3 - Training objectives for each age group**<br>
**4 - The success rate of each training program in achieving its objective**

```python
exercise_goal_relation = pd.crosstab(df['workout_module_used'], df['goal'])

plt.figure(figsize=(12, 7))
sns.heatmap(exercise_goal_relation, annot=True, fmt='d', cmap='Blues')
plt.title('Relationship between Workout Type and User Goals')
plt.ylabel('Workout Module')
plt.xlabel('User Goal')
plt.show()
```

```python
pivot = pd.crosstab(df['gender'], df['goal'])
plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt='d', cmap='Blues')
plt.title('Gender vs Goal')
plt.yticks([1.5,.5], ['Male','Female'])
plt.xlabel('user Goal')
plt.ylabel('Gender')
plt.show()
```

```python
pivot = pd.crosstab(df['Age Group'], df['goal'])
plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt='d', cmap='Blues')
plt.title('Age vs Goal')
plt.xlabel('user Goal')
plt.ylabel('Age Group')
plt.show()
```

```python
workout_goal_rate = df.groupby('workout_module_used')['goal_reached_numeric'].mean() * 100

plt.figure(figsize=(10, 5))
workout_goal_rate.sort_values().plot(kind='bar', color='#457B9D')
plt.title('Goal Success Rate by Workout Type (%)')
plt.ylabel('Percentage %')
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate(workout_goal_rate):
    plt.text(i, v + 0.5, f"{v:.1f}", ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

**1 - Average calories burned per workout**<br>
**2 - Average workout duration for each training program**<br>
**3 - Average workout time by age group**<br>
**4 - Average workout duration for men and women**

```python
avg_calories = df.groupby('workout_module_used')['calories_burned'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 6))
avg_calories.plot(kind='bar', color="#457B9D")
plt.title('Average Calorie Burn Intensity', fontsize=15)
plt.xlabel('Workout Type', fontsize=12)
plt.ylabel('Calories Burned per Minute', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate(avg_calories):
    plt.text(i, v + 0.5, f"{v:.1f}", ha='center', fontweight='bold')
plt.show()
```

```python
avg_duration = df.groupby('workout_module_used')['session_duration_minutes'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
avg_duration.plot(kind='bar', color="#457B9D", edgecolor='black')
plt.title('Average Duration per Workout Type', fontsize=15)
plt.xlabel('Workout Module', fontsize=12)
plt.ylabel('Average Time (Minutes)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate(avg_duration):
    plt.text(i, v + 0.5, f"{v:.1f}", ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
duration_by_age = df.groupby('Age Group')['session_duration_minutes'].mean()

plt.figure(figsize=(8, 5))
duration_by_age.plot(kind='bar', color='#457B9D', edgecolor='black')
plt.title('Average Workout Duration by Age Group', fontsize=15)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Average Duration (Minutes)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(duration_by_age):
    plt.text(i, v + 0.5, f"{v:.2f}", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
duration_by_gen = df.groupby('gender')['session_duration_minutes'].mean()

plt.figure(figsize=(8, 5))
duration_by_gen.plot(kind='bar', color='#457B9D', edgecolor='black')
plt.title('Average Workout Duration by Gender', fontsize=15)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Average Duration (Minutes)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate(duration_by_gen):
    plt.text(i, v + 0.5, f"{v:.2f}", ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
daily_sessions = df.groupby('session_date').size()

plt.figure(figsize=(12, 5))
daily_sessions.plot(color='#457B9D', linewidth=2)
plt.title('Daily Training Sessions Trend', fontsize=14)
plt.ylabel('Number of Sessions')
plt.grid(True, alpha=0.3)
plt.show()
```
